```python
# "%" = Magic Command
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
fig = plt.figure()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker='D', color='green', linewidth=2, label='Boston')
#Change the line color to green.
#Change the marker to a diamond.
#Change the line width to 2 points.
#Add a y-axis limit.
ax.set_ylim(0, 45)
#Add a title and axes labels.
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#Add grid lines.
ax.grid()
#Add a legend for the city of Boston.
ax.legend()
```

```python
#Switch the axis so that the Fare($) data is on the x-axis.
#Change the bar color to cyan.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
#Add a legend for the city of Chicago.
ax.legend()
#Add a title and axes labels.
ax.set_title('PyBer Fare by Month')
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
#Invert the y-axis data.
ax.invert_yaxis()
```

```python
#Create a Scatter Plot Using the Object-Oriented Interface
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#Create a Bubble Chart Using the Object-Oriented Interface
```

```python
#Bubble chart with relative size on each plot
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], c='m', edgecolors='black', label='Boston')
#Change the color of the markers to sky blue.
#Change the size of the markers to 5 times each data point.
#Make the color 20% transparent.
#Add a black edge color to the circles.
#Make the linewidth of the circles 2 points.
#Add a legend for the city of Boston.
ax.legend()
#Add a title.
ax.set_title('PyBer Fare by Month')
#Switch the axis so that the Fare($) data is on the x-axis.
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
#Add a limit for the x-axis data.
ax.set_xlim(0, 45)
#Invert the y-axis so that January is at the top.
ax.invert_yaxis()
```

```python
#Create a Pie Chart Using the Object-Oriented Interface
```

```python
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
#Add new colors of your own choosing or use the colors from the previous pie chart.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots()
ax.pie(y_axis,
      explode=explode_values,
      colors=colors,
      labels=x_axis,
      autopct='%.1f%%',
      shadow=True,
      center=(8,8))

plt.show()
#Add a percentage to one decimal place to each wedge of the pie.
#Increase the figure size to 8x8.
#Explode the two highest percentage months.
#Add a shadow.
#Add a start angle so that January is at the top.
#Reverse the order so that the month order is in a clockwise direction.
```

```python
#Add Error Bars to a Line Chart - Object-Oriented Interface
```

```python
import statistics
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#Add Error Bars to a Bar Chart - Object-Oriented Interface
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#Change the Major Ticks - Object-Oriented Interface
```

```python
import numpy as np
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
#Change the Minor Ticks - Object-Oriented Interface
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
